2. Never starts with a number
3. A variable name can only contain alpha-numeric characters and underscores
4. All are case sensitive
5. A variable name cannot be a python key word

# MultiWord Variabels
**Camel Case**
Camel case is a common convention for naming variables in Python. It involves writing the first word in lowercase and capitalizing the first letter of each subsequent word. For example, "myVariableName" is an example of camel case. This convention helps improve readability and makes it easier to understand the purpose of the variable at a glance.

**Snake Case**
Snake case is another common convention for naming variables in Python. It involves writing all words in lowercase and separating them with underscores. For example, "my_variable_name" is an example of snake case. This convention is often used in Python to improve readability and make it easier to understand the purpose of the variable.


**Pascal Case**

pascal case is a convention for naming variables in Python where the first letter of each word is capitalized, and there are no spaces or underscores between the words. For example, "MyVariableName" is an example of Pascal case. This convention is often used for naming classes in Python.



```python
# Variable naming conventions
# camel case
myVariableName = "This is an example of camel case."
print(myVariableName)

# snake case
my_variable_name = "This is an example of snake case."
print(my_variable_name)

# pascal case
MyVariableName = "This is an example of Pascal case."
print(MyVariableName)
```

# Global Variables 
Global variables are defined outside of any function and can be accessed from anywhere in the code. Local variables are defined inside a function and can only be accessed within that function.


```python
x = "awesome"

def myfunc():
    print("python is " + x)
    
myfunc()     # calling the function to execute the code inside it.
        
        
```

# Local variable
