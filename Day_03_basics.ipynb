# Numbers and Casting
Concept: Python is smart, but sometimes it needs help. Casting is simply telling Python, "Treat this value as a different type
 
 `Casting in python is used to change data type`

```python
age = "25" # This is a string (Text)
print(type(age))

# If you try: print(age + 1) -> It will crash (TypeError).


# We need to CAST it to an integer first:
actual_age = int(age)
print(type(actual_age))
print(actual_age + 1)
```

# Python Data Types

In Python, there are several built-in data types that are used to store and manipulate data.Data Types are the atomic units of information (the "building blocks"). int, float, str, and bool are just basic pieces of data. They represent a single "thing."


 Here are some of the most common ones:

1. **Numeric Types**:
   - `int`: Represents integer values (e.g., 1, 42, -5).
   - `float`: Represents floating-point numbers (e.g., 3.14, -0.001).
   - `complex`: Represents complex numbers (e.g., 2 + 3j).
2. **Text Type**:
   - `str`: Represents a sequence of characters (e.g., "Hello, World!).
3. **None Type**:
   - `NoneType`: Represents the absence of a value (e.g., `None`).
4. **Boolean Type**:
   - `bool`: Represents a value that can be either `True` or `False`.


These data types allow you to store and manipulate different kinds of data in Python, making it a versatile programming language for various applications.   



# Python Data Structures
Data Structures are the containers built to organize those atoms. A list, dict, or set is a structure that holds those atoms together in a specific way

1. **Sequence Types**:
    - `list`: An ordered, mutable(changable) collection of items (e.g., [1, 2, 3], ['a', 'b', 'c']).
    - `tuple`: An ordered, immutable collection of items (e.g., (1, 2, 3), ('a', 'b', 'c')).
    - `range`: Represents a sequence of numbers (e.g., range(0, 10)).
2. **Mapping Type**:
   - `dict`: Represents a collection of key-value pairs (e.g., {'name': 'Alice', 'age': 30}).
3. **Set Types**:   
    - `set`: An unordered collection of unique items(it deletes the duplicates) (e.g., {1, 2, 3}).
    - `frozenset`: An immutable version of a set (e.g., frozenset({1, 2, 3})).

                                                            
                                                            


```python
# 1. A list for shelf order

shelf = ["Atomic Habits", "Deep Work"]

#.append() adds a new book to the end of the shelf, we use append in list because we are adding a new book to the end of the shelf, and we want to maintain the order of the books on the shelf  
shelf.append("The Psychology of Money") 




# 2. A set for unique IDs
book_ids = {101, 102}

# Try adding 101 again!, it will not add a new element to the set because sets do not allow duplicate values, so it will ignore the duplicate value and keep the set as is
book_ids.add(101) 


# 3. A dict for Title -> Author mapping
library = {"Atomic Habits": "James Clear"}

#adding a new key-value pair to the dictionary, we use the key "Deep Work" and the value "Cal Newport" to add a new book to the library, and it will maintain the mapping of book titles to authors in the dictionary
library["Deep Work"] = "Cal Newport" 


#check that if dict allow another key with the same name, it will update the value instead of adding a new key-value pair
library["Atomic Habits"] = "James Clear Updated"#if we add a new key-value pair with the same key, it will update the value instead of adding a new key-value pair




print(f"Shelf: {shelf}")           # Print the list of books on the shelf
print(f"IDs: {book_ids}")          # Print the set of unique book IDs
print(f"Library: {library}")       # Print the dictionary mapping book titles to authors
```

# Python Strings
**The Master Class**

let's look at the most used Atomic Type in the real world: str (Strings).
1. In Python, a string is a sequence. This means you can "slice" it just like a list.
2. Length : How many characters are in string
3. Membership: that if a letter exists or not 



```python
name = "Alice is a good programmer"

#slicing the string to get the first 5 characters, we use slicing to get the first 5 characters of the string, and it will return "Alice"

print(name[0:5])
print(name[:5]) # This is the same as name[0:5]

# finding length
print(len(name))

# checking membership
print("y" in name)
```

**Slicing in real-world**
1. Data cleaning: Removing unwanted prefixes (e.g., getting the phone number out of "+92-300-1234567" by slicing the first 3 characters)

2. Validation: Checking the last few letters of a filename to see if it's a photo (".jpg") or a document (".pdf").

3. Text Analysis: Extracting initials from a name ("Mirha Nem" -> M and N).


```python
# A challange

file_name = "data_report_2026.csv"
print(file_name[12:16]) # This will print "2026" because it will slice the string from index 12 to index 16
print(file_name[-4:]) # This will print ".csv" because it will slice the last 4 characters of the string
```

The easier way to slice a string

split by a character, we use the split() method to split the string by the underscore character, and it will return a list of strings that are split by the underscore character

```python
file_name = "data_report_2026.csv"

parts = file_name.split("_")
print(parts) # This will print ['data', 'report', '2026.csv']


#Now just grab the year
year = parts[2].split(".")[0] # This will split the string "2026.csv" by the dot character and return the first part which is "2026"
print(year) # This will print "2026"
```

# Python Boolean and operators

Category,Operators & What they do

1. Arithmetic,"+, -, *, /, %","Math (Note: % is ""Modulus""—it gives you the remainder).

2. Comparison,"==, !=, >, <, >=, <=",Asking questions; returns a bool (True/False).

3. Logical,"and, or, not","Combining conditions (The ""Logic Gate"")."

```python
print(type(10 > 5))   
print(type(10 == 5))
print(type("apple" == "apple"))


# it will result in boolean values, True or False, and the type of these values is <class 'bool'>, because the comparison operators (>, ==) return boolean values indicating whether the comparison is true or false.
```

```python
# Alice
age1 = 20
has_id1 = True
can_enter1 = (age1 >= 18) and has_id1

# Ali
age2 = 23
has_id2 = False
can_enter2 = (age2 >= 18) and has_id2

print(f'Alice can enter: {can_enter1}') # True
print(f'Ali can enter: {can_enter2}')   # False
```

**The fascinating fact**

When you run dir(5), you’ll see a list of functions (like __add__, __sub__). This reveals the secret: When you type 5 + 5 in Python, you are actually calling a hidden function inside the number 5.

You aren't just doing math; you are sending a command to an object. This is the foundation of Object-Oriented Programming (OOP), which makes Python one of the most powerful languages in the world. Most beginners think 5 is just 5, but 5 is a small, capable, active object.

```python
# See all the "superpowers" of an integer
print(dir(5))

# See all the "superpowers" of a list
print(dir([]))
```
