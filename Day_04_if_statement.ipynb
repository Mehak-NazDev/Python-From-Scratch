# Pyhthon Elif Statements

The keywoed "elif" in python means that if previous condition is not true then execute and try this block of code

```python
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

# Pyhton Else
complete code

else: The "Default" path if no other conditions are met

```python
# The "Catch-all"
score = 45

if score >= 90:
    print("Grade: A")
elif score >= 70:
    print("Grade: B")
else:
    print("Grade: F (Needs improvement)")
```
