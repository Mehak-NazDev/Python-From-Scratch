**Comments in pyhton**

A single line comment starts from " # ". The line statement with # will not be printed. 
A multiline comment starts from """ and ends at """. It serves the same purpose as single line comment.

You can use these for

1. Readability
2. To note something  
3. To help yourself to mark something important   
4. Add a heading to understand the code

**Case Sensitivity in python**

Python is a case-sensitive language, which means that variables with the same name but different cases are treated as different variables. For example, "Mehak" and "mehak" are two different variables in Python.


# Python Variables

Variables are containers for storing data values. Variables are created the moment we assign it a value.
A variable can store values of following data types
1. Integer
2. Float
3. String 
4. Boolean
5. Mapping 
6. Sequence (list, tuple, range)

**\n in python**

\n is used to print each variable on a new line, it breaks the line after each variable is printed.

```python
x = 4  #integer
y = 3.14 #float
name = "Mehak" #string  
is_valid = True #boolean
dict_var = {"key": "value"} #dictionary
list_var = [1, 2, 3] #list
tuple_var = (1, 2, 3) #tuple
range_var = range(5) #range

# Print the variables
print (f" x: {x} \n y: {y} \n name: {name} \n is_valid: {is_valid} \n dict_var: {dict_var} \n list_var: {list_var} \n tuple_var: {tuple_var} \n range_var: {range_var}")
```

# Practical Problems on Variables
The type mismatch error occurs when you try to perform an operation on a variable that is not compatible with the operation. 
For example, if you try to add a string and an integer, you will get a type mismatch error. 

**Solutions**
1. Type-Casting 

typecasting age to string using str() function. In simple words, we are converting the integer age to a string using the str() function so that we can concatenate it with the string name.

2. f strings
 f strings: we can also use f strings to print the variables in a more readable way.that here we are not treating age as a string but as an integer and we are using f strings to print the variables in a more readable way.


```python
name = "Ali"
age = 20
#uncomment the below line to see the error
# print (name + age)  Notice the typeError : can only concatenate str (not "int") to str
#corrected code:
# 1.
print (name + str(age))  
# 2. 
print (f"Hi, I am {name}, and I am {age} years old.")
print (f"Hi, I am {name}, and I am {age} years old. My name has {len(name)} characters.") #len() is a function that returns the length of the string
```

```python
name = "Maha"
name = "mirha"
# printing name variable will give output as "mirha" because the last assignment to the variable name is "mirha" because the previous assignment to the variable name is overwritten by the last assignment to the variable name.
print(name)

name = "Ali"
Name = "Ahmad"

print (name, Name) # python is case sensitive, so name and Name are two different variables.
```

 Notice the difference between the above two lines of code, the first line will concatenate the two strings "20" and "30" to give "2030" while the second line will add the two integers 20 and 30 to give 50. Why is this occuring because the first line is treating the numbers as strings while the second line is treating the numbers as integers. This is because the first line is using the str() function to convert the integers to strings before concatenating them, while the second line is not using any function to convert the integers to strings, so it treats them as integers and adds them together.


```python
print (str(20) + str(30))
print (20 + 30)
```
