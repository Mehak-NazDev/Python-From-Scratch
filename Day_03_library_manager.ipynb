# Day 3: Practical Project - Library Manager

**Objective:** Combine Data Types, Structures, and Functions to build a simple library that handles 3 real-world "edge cases."

**Concepts Applied:**

- **Nested Dictionaries:** Using a dictionary to store complex book data.
- **Data Sanitization:** Using `.strip()` and `.title()` to clean messy user input.
- **Validation Gates:** Using `if` conditions to prevent invalid data from entering our database.

### The "Gatekeeper" Logic
Before adding a book, our function acts as a gatekeeper to ensure:
1. The title isn't just empty spaces.
2. The title is formatted consistently (to avoid "atomic habits" vs "Atomic Habits").
3. We don't accidentally overwrite existing books without intending to.

```python
# The Database
library = {}

def add_book(title, author, status):
    # EDGE CASE 1: Handle Case Sensitivity (Store everything in Title Case)
    clean_title = title.strip().title()
    
    # EDGE CASE 2: Prevent empty titles or invalid input
    if not clean_title or len(clean_title) < 2:
        print("Error: Please provide a valid book title.")
        return # Stop the function here

    # EDGE CASE 3: Check if it already exists
    if clean_title in library:
        print(f"Update: '{clean_title}' already exists. Updating status.")
        library[clean_title]["Status"] = status
    else:
        # Success: Add new entry
        library[clean_title] = {"Author": author.strip(), "Status": status}
        print(f"Success: '{clean_title}' added to library.")

# Testing the edge cases
add_book("atomic habits", "James Clears", "Read")      # Case sensitivity check
add_book("  ", "Unknown", "Unread")                   # Empty title check
add_book("Atomic Habits", "James Clear", "Reading")   # Already exists check
add_book("The Great Gatsby", "F. Scott Fitzgerald", "Unread")  # Valid entry 

print("\nYour Final Library: ")
print(library)

```

### Research Findings & Edge Cases
- **Case Sensitivity:** By using `.title()`, we ensure that the first letter of each word is capitalized, which helps maintain consistency in the database. This way, "atomic habits" and "Atomic Habits" will be treated as the same book, preventing duplicates and ensuring that the library remains organized.

- **Empty Inputs:** Used `if not clean_title` to catch empty strings or white-space-only entries.

- **Persistence:** Realized that variables like `library` act as our temporary database during the program's runtime.


**Library with correct year format**  

```python
def add_book_with_year(title, author, year):
    # We cast to int right here to ensure the data is "clean"
    year_int = int(year) 


#Adding an edge case to check if the year is a valid integer
    if not isinstance(year_int, int) or year_int < 0:
        print(f"Error: {year} is not a valid publication year.")
        return 
    library[title] = {"Author": author, "Year": year_int}
    print(f"Added {title}, published in {year_int}")
    
   
    

# Calling it
add_book_with_year("Deep Work", "Cal Newport", "2016")
add_book_with_year("Hamlet", "William Shakespeare", "-1603")  # Invalid year check

```

**one key note:**

Since we are casting year_int = int(year), if a user inputs a letter (like "abc"), the code will crash before it reaches if statement so for that we can use try and except block to catch the error and print a message instead of crashing the program.we will learn about try and except block in the later lesson.
